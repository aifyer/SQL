### Module 2 -- Relational Databases

**Entity-Relationship Model**

- Data is stored in tables
- Allows for data independence
- Biulding blocks are Entities and Attributes
- For eg. The Entity "Book" can have Atributes like "title","edition","year"
- Entity becomes a table, and Attributes becomes columns of the table
- Each Attributes carry different data format like: number, character, dates, currency ...
- Each table is assigned a 'Primary Key' which is a unique id for each row.
- 'Foreign Key' is the Primary Key of other tables that provide a link between the tables.
- Rows of data values are called 'tuples'


**Types of SQL Statements**

_DDL(Data Definition Language statements)_
- DDL is used to define, change, or drop database objects such as tables.
- Common DDL statements:
  - CREATE: create tables and definding columns.
  - ALTER: alter tables including adding and droping columns and modify their data types.
  - TRUNCATE: deleting data in a table, but not the table itself.
  - DROP: deleting tables.
    
_DML(Data Manipulation Language statements)_
- DML is used to read and modify data and tables.
- Also refered to as CRUD operations (Create, Read, Update & Delete rows)
- Common DML statements:
  - INSERT: insert rows into a table
  - SELECT: read and selecte rows
  - UPDATE: edit row 
  - DELETE: remove rows

### Module 3 -- Intermediate SQL

#### Intermediate cmd
**Retrieve data using string patterns, ranges, values**

- _string pattern_: Select rows with regular expression like name start with letter 'R', use the `like` and wildcard '%' which can be any location:
```sql
select firstname from Author WHERE firstname like 'R%'
```
- _range_: Select rows with pages between 290 and 300. there are two ways, the second is inclusive, exactly the same as the first. 
```sql
select title,pages from Book WHERE pages >=290 AND pages <=300;
select title,pages from Book WHERE pages between 290 and 300;
SELECT * FROM EMPLOYEESWHERE (SALARY BETWEEN 60000 AND 70000) AND DEP_ID = 5;
```
- _values_:
```sql
select firstname,lastname,country from Author WHERE country='AU' OR country='BR';
select firstname,lastname,country from Author WHERE country IN ('AU','BR','CA');
```

**Sorting result sets**

- `ORDER BY` :sorting ascending default. use DESC if want reverse.
- Can also specify the column sequence number in the select statement(not original table).
```sql
select title from Book ORDER BY title;
select title from Book ORDER BY title DESC;
select title,pages from Book ORDER BY 2;
SELECT F_NAME, L_NAME, DEP_ID FROM EMPLOYEES ORDER BY DEP_ID DESC, L_NAME DESC;
```

**Grouping result sets**

- Eliminating Duplicates: `DISTINCT`.
```sql
select distinct(country) from Author;
```
- Group by country and count the numbers of author in that country.
- column name for the count will be the new col number. can specify it by `as`
```sql
select country,count(country) from Author GROUP BY country;
select country,count(country) as Count from Author GROUP BY country;
SELECT DEP_ID, COUNT(*) AS "NUM_EMPLOYEES", AVG(SALARY) AS "AVG_SALARY" FROM EMPLOYEES 
    GROUP BY DEP_ID HAVING count(*) < 4 ORDER BY AVG_SALARY;
```
- _Restricting result_: `having`, only show the count above 4, it works only with group by clause.
```sql
select country,count(country) as Count from Author 
    GROUP BY country
    having count(country)>4;
```  
- Retrieve from two tables together
```sql
SELECT D.DEP_NAME , E.F_NAME, E.L_NAME
FROM EMPLOYEES as E, DEPARTMENTS as D
WHERE E.DEP_ID = D.DEPT_ID_DEP
ORDER BY D.DEP_NAME, E.L_NAME DESC;
```

#### Built-in Functions

**Aggregate Funcitions**

- Including `SUM()`, `MIN()`, `MAX()`, `AVG`..
- CAN NOT been evaluated in a WHERE clause.
```sql
select SUM(COST) as SUM_OF_COST from PetTable;
select MAX(QUANTITY) from PetTable;
select MIN(ID) from PetTable WHERE ANIMAL = 'Dog';
select AVG(COST) from PetTable;
```
- Can do calculation between columns
```sql
select AVG(COST/QUANTITY) from PetTable WHERE ANIMAL ='Dog';
```

**Scalar and String Functions**

- Including: `ROUND()`, `LENGTH()`, `UCASE()`, `LCASE()`
- Can be use in where clause

```sql
select ROUND(COST,2) from PetTable;
select LENGTH(ANIMAL) from PetTable;
select UCASE(ANIMAL) from PetTable;
select * from PetTable where LCASE(ANIMAL)='cat';
select DISTINCT(UCASE(ANIMAL)) from PetTable;
```

**Date and Time Functions**

- SQL have DATE, TIME, TIMESTAMP data type:
  - DATE: `YYYYMMDD`
  - TIME: `HHMMSS`
  - TIMESTAMP: `YYYYXXDDHHMMSSZZZZZZ`, here XX is month, and ZZZZZ is micro-seconds
- Date/Time functions include: `YEAR()`,`MONTH`,`DAY`,`DAYOFMONTH`,`DAYOFWEEK`,`DAYOFYEAR`,`WEEK`,`HOUR`,`MINUTE`,`SECOND`.
- They can be used in the `where` clause
- They can also do Time Arithmetic `DATE_ADD`
- Special Registers: `CURRENT_DATE`, `CURRENT_TIME`
- Other Date Functions: `DATE_ADD`, `DATE_SUB`, `DATEDIFF`
- `FROM_DAYS` : take the number of days and return the required formatted output
  
```sql
select DAY(RescueDate) from PetTable where ANIMAL = 'cat';
select COUNT(*) from PetTable where MONTH(RescueDate)='05';
select DATE_ADD(RescueDate, INTERVAL 3 DAY) from PetTable;
select FROM_DAYS(DATEDIFF(CURRENT_DATE,RescueDate)) from PetTable;
SELECT DATE_ADD(RESCUEDATE, INTERVAL 2 MONTH) FROM PETRESCUE;
select (RESCUEDATE + 3 DAYS) from PETRESCUE;
SELECT DATE_SUB(RESCUEDATE, INTERVAL 3 DAY) FROM PETRESCUE;
select (CURRENT DATE - RESCUEDATE) from PETRESCUE;
SELECT ID, DATE_ADD(RESCUEDATE, INTERVAL 1 YEAR) FROM PETRESCUE;
```

#### SUB-Queries and Nested Selects

- Sub-query: A query inside another query  
```sql
select Column1 from theTable where column2=(select MAX(column2) from theTable);
```

- use in a `where` clause
- use in column expressions
- use in FROM clause to have derived tables or table expressions, powerful when working with multiple tables.
- To retrieve the list of employees who earn more than the average salary:
```sql
select F_NAME,L_NAME,Salary from theTable where salary < (select AVG(salary) from theTable);
select EMP_ID, SALARY, (select AVG(SALARY) from employees) as AVG_SAL from employees;
select * from ( select EMP_ID, F_NAME, L_NAME, DEP_ID from employees) AS EMP4ALL;
```

#### Multiple Tables
- sub-queries
```sql
select * from employees where depId In (select depId from departments);
select * from employees where depId In (select depId from departments where loc_id = 'L0002');
```
- implicit Join.
```sql
--- Cross join/ full join:
select * from employees, departments;
--- Inner join:
select * from employees E, departments D where E.depId = D.depId;
select E.empID, D.depName from employees E, departments D where E.depId = D.depId;
```

### Module 6 -- Advanced SQL

**Views**

- Views include specified columns from multiple base tables and existing views
- Views can be queried like a table
- Only the definition of the view is stored, not the data
- The CREATE VIEW statement creates a view based on one or more tables or views
- Views can
  - show a selection of data for a given table
  - combine two or more tables in a meaningful ways
  - simplify access to data
  - show only portions of data in the table

_create a view_
```sql
CREATE VIEW view_name colname1,colname2 
AS SELECT colname1,colname2
FROM table_name
WHERE predicate;
```
_Update a view_
```sql
CREATE OR REPLACE VIEW EMPSALARY AS
SELECT EMP_ID, F_NAME, L_NAME, B_DATE, SEX, JOB_TITLE,
MIN_SALARY, MAX_SALARY
FROM EMPLOYEES, JOBS
WHERE EMPLOYEES.JOB_ID = JOBS.JOB_IDENT;
```

_select from a view and drop a view_
```sql
select * from view_name;
drop VIEW view_name;
```

**Stored Procedures**
- A set of SQL statements stored and executed on the database server
  - write in many different languages
  - accept information in the form of parameters
  - return results to the client
- Benefits of stored procedures
  - reduction in network traffic
  - improvement in performance
  - reuse of code
  - increase in security

```sql
DELIMITER $$ --change delimite to $$. means code end in the next $$
CREATE PROCEDURE UPDATE_SAL (IN empNUM CHAR(6), IN rating SMALLINT) --input empNUM and rating
BEGIN
UPDATE employees SET salary = salary * 1.10 WHERE emp_id = empNum AND rating = 1;
UPDATE employees SET salary = salary * 1.05 WHERE emp_id = empNum AND rating <> 1;
END
$$
DELIMITER ; --change delimiter back to ;
```

- stored procedure can be called from external applications and Dynamic SQL statements
```sql
CALL UPDATE_SAL ('E1001',1)
```

**ACID transactions**
- Indivisible unit of work
- Consists of one or more SQL statements
- Either all happens or none
  - Atomic: all changes must be performed successfully or not at all
  - Consistent: Data must be in a consistent state before and after the transaction
  - Isolated: No other process can change the data while the transaction is running
  - Durable: The changes made by the transaction must persist
- can be called by other languages
```sql
BEGIN
UPDATE Table SET a = a-1;
UPDATE Table SET b =b-2;
COMMIT --or ROLLBACK(if any statement failure)
```
