#### 2. Create the alter_key procedure and upload result tables

```python
%%bash

CREATE OR REPLACE PROCEDURE add_keys(runID INTEGER)
LANGUAGE plpgsql
AS $$
BEGIN
    -- result table
    EXECUTE format('
        ALTER TABLE result_run%s
        ADD PRIMARY KEY (id),
        ADD FOREIGN KEY (run_id) REFERENCES run_info(run_id),
        ADD FOREIGN KEY (water_grade_id) REFERENCES water_grade(id),
        ADD FOREIGN KEY (additive_id) REFERENCES additive(id)',
        runID);
        
    -- summary table
    EXECUTE format('
        ALTER TABLE summary_run%s
        ADD FOREIGN KEY (run_id) REFERENCES run_info(run_id),
        ADD FOREIGN KEY (additive_id) REFERENCES additive(id)',
        runID);

END;
$$;
```

```python
# define a function to upload the result_run, summary_run table, and run_record to run_info
def upload_experiment_run(runID: int, data_path: str, db_engine: Engine):  # ✅ Engine not engine
    """
    Load CSV data for a given runID and upload it into optph_db.
    """
    # -----------------------------
    # Build paths
    # -----------------------------
    run_path        = os.path.join(data_path, f"run{runID}")
    result_file     = os.path.join(run_path, f"result_run{runID}.csv")
    summary_file    = os.path.join(run_path, f"summary_run{runID}.csv")
    record_file     = os.path.join(run_path, f"Record_Form_experiment_run{runID}.csv")
    out_record_file = os.path.join(run_path, f"record_run{runID}.csv")

    # -----------------------------
    # check results file
    # -----------------------------
    df_results = pd.read_csv(result_file)
    df_sum = pd.read_csv(summary_file)

    eg_results = pd.read_csv(os.path.join(data_path, "result_run_template.csv"))
    eg_sum = pd.read_csv(os.path.join(data_path, "summary_run_template.csv"))

    if list(df_results.columns) != list(eg_results.columns):
        raise ValueError(f"result_run{runID}.csv columns do not match template")

    if list(df_sum.columns) != list(eg_sum.columns):
        raise ValueError(f"summary_run{runID}.csv columns do not match template")

    # -----------------------------
    # Process record file
    # -----------------------------
    additive_pairs = (
        df_results[['additive_id', 'add_final_concentration_M']]
        .dropna() 
        .drop_duplicates()
        .sort_values(['additive_id', 'add_final_concentration_M'])
        .reset_index(drop=True)
    )
    additive_json = [
        {
            "additive_id": int(row['additive_id']),
            "concentration": float(row['add_final_concentration_M'])
        }
        for _, row in additive_pairs.iterrows()
    ]
    df_record     = pd.read_csv(record_file)
    df_horizontal = df_record.set_index('records')['value'].to_frame().T
    df_horizontal['additive_info'] = json.dumps(additive_json)
    df_horizontal.to_csv(out_record_file, index=False)

    # -----------------------------
    # PostgreSQL connection
    # -----------------------------
    conn = psycopg2.connect(
        dbname="optph_db",
        user="liuh339",
        host="localhost",
        port="5432"
    )
    cur = conn.cursor()

    try:
        # -----------------------------
        # Step 1: Load run_info
        # -----------------------------
        with open(out_record_file, "r") as f:
            cur.copy_expert(
                "COPY run_info FROM STDIN WITH CSV HEADER",
                f
            )

        # -----------------------------
        # Step 2: Upload result & summary (auto-create tables)
        # -----------------------------
        df_results.to_sql(
            f"result_run{runID}", 
            db_engine, 
            if_exists="fail",  
            index=False
        )

        df_sum.to_sql(
            f"summary_run{runID}", 
            db_engine, 
            if_exists="fail",  
            index=False
        )

        # -----------------------------
        # Step 3: Add keys
        # -----------------------------
        cur.execute("CALL add_keys(%s);", (runID,))
        conn.commit()

    except Exception as e:
        conn.rollback()
        raise RuntimeError(f"Failed to upload run {runID}: {e}")

    finally:
        cur.close()
        conn.close()
```

```python
#upload each individual run results

run_ids = [1,2,3,5,6,9,14,17]
engine = create_engine("postgresql+psycopg2://liuh339@localhost:5432/optph_db")

for runID in run_ids:
    print(f"Processing run {runID}...")
    upload_experiment_run(
    runID     = runID,
    data_path = "/Users/liuh339/Library/CloudStorage/OneDrive-PNNL/Documents/projects/Yangang/optph_db_files",
    db_engine    = engine
)
    print(f"run {runID}: Done")
```
